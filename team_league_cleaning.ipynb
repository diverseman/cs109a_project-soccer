# FIFA player data cleaning

This notebook will process the FIFA Team and League CSV and clean/modify so they are ready for analysis & learning.

```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup
```

```python
file_name = 'teams_and_leagues.csv'
input_directory = 'data/'
output_directory = 'cleaned_data/'
```

## cleandata_team_id
This function will take a DataFrame and find the team IDs each team and categorize them, allowing us to see which league a team is part of.

```python
def find_team_id(df):
    
    df = df.copy(deep=True)
    id_list = df.url.values
    team_list = []
    
    for i in id_list:
        url = "http://sofifa.com/team/" + str(i) + "/?"
        home_page = requests.get(url)
        soup = BeautifulSoup(home_page.text, "html.parser")
        webpage_title = soup.title.text
        
        regex = ".+(?= FIFA)"
        
        print("Processing: Team ID:",i)
        team_name = re.findall(regex, webpage_title)[0]
        team_list.append(team_name)
    df["team"] = team_list
        
    return df
```

## Main logic for the cleaning

```python
input_path = input_directory + file_name
df = pd.read_csv(input_path)
    
df = find_team_id(df)

output_path = output_directory + file_name
df.to_csv(output_path)
print('\nCleaned file written.')
```
