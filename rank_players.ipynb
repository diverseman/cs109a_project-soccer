取数据

```python
x=fifa_19[useful_features]
y=fifa_19.overall
```

```python

from sklearn.ensemble import RandomForestRegressor
model_1=RandomForestRegressor()
model_1.fit(x,y)
```

```python
from sklearn.metrics import mean_squared_error as mse 
x_test,y_test=fifa_20.loc[fifa_20.sofifa_id.isin(fifa_19.sofifa_id)][useful_features],fifa_20.loc[fifa_20.sofifa_id.isin(fifa_19.sofifa_id)].overall
y_pre=model_1.predict(x_test)
error= mse(y_test,y_pre)
print(f'this is model_1`s mse {error}')
```

调整randomForest的参数

```python
model_2=RandomForestRegressor(oob_score=True,max_depth=100,n_estimators=50).fit(x,y)
y_pre_model_2=model_2.predict(x_test)
error_1_2=mse(y_test,y_pre_model_2)
print(f'this is the model_1_2 mse(changed):{error_1_2}')
```

```python
team =['FC Barcelona',
'FC Bayern München',
'Real Madrid',
'Paris Saint-Germain',
'Juventus' ,
'Manchester City',
'Liverpool FC']
```

```python
pre_team=fifa_20.loc[map(lambda x: team.__contains__(x),fifa_20.club)]
x_pre_team,y_pre_team=pre_team[useful_features],pre_team.overall
model_pre_team=RandomForestRegressor(oob_score=True,max_depth=100,n_estimators=50).fit(x,y)
y_pre_team_model=model_pre_team.predict(x_pre_team)
error_pre_team=mse(y_pre_team,y_pre_team_model)
print(f'this is the model_pre_team mse:{error_pre_team}')
```

```python
pre_team=fifa_20.loc[map(lambda x: team.__contains__(x),fifa_20.club)]
x_test_team,y_test_team=pre_team[useful_features],pre_team.overall
model_pre_team=RandomForestRegressor(max_depth=100,n_estimators=50).fit(x,y)
y_pre_team_model=model_pre_team.predict(x_test_team)
error_pre_team=mse(y_test_team,y_pre_team_model)
print(f'this is the model_pre_team mse:{error_pre_team}')
```

```python
pre_team=fifa_20.loc[map(lambda x: team.__contains__(x),fifa_20.club)]
x_test_team,y_test_team=pre_team[useful_features],pre_team.overall
model_pre_team=RandomForestRegressor(max_depth=100,n_estimators=150).fit(x,y)
y_pre_team_model=model_pre_team.predict(x_test_team)
error_pre_team=mse(y_test_team,y_pre_team_model)
print(f'this is the model_pre_team mse:{error_pre_team}')
```
